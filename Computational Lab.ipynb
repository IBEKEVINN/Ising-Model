### **Experiment 1: Magnetisation versus Temperature for small lattice**

##### **Question 1: Describe what is happening physically as you increase T in the 'cold' start lattice inexperiment 1, and how your plot of magnetization versus temperature reflects that physical change.**

```python
## Cold start for 30x30 lattice

exp1 = model.plots(N=30)
exp1.mag()
exp1.show()
exp1.lattice()
exp1.show()
```

When we initially start in the 'cold' lattice in experiment 1, all the spins are aligned - resulting in 'full' magnetisation (as magnetisation is defined as the average spin). We note that thermal energy in the system will tend to interfere with spin alignment by spontaneously flipping spins. As such, it is evident in the plot that as temperature increases, we have some electron spins that are being flipped randomly. As we assume a ferromagnetic material for this model, spins will tend to align, resulting in these flipped spins to flip back to its original. 

As such, there is a balance between the temperature of the system spontaneously flipping spins and the ferromagnetic property of the lattice reversing these flips. As temperature increases, this balance shifts and the number of randomly flipped spins exceeds the lattice's ability to align these flips to its neighbours - resulting in a chaotic structure that has no net magnetisation. The point at which the lattice is unable to retain its ordered structure is its critical temperature (somewhere between 2 and 2.5 J/kb).

##### **Describe the differences between the magnetization versus temperature plots for the 'cold' start and the 'hot' start. What in the program might be causing those differences?**

```python
## High start for 30x30 lattice

exp1 = model.plots(N=30, start='High')
exp1.mag()
exp1.show()
exp1.lattice()
exp1.show()
```

Alternatively, in the 'hot' start, all the spins are randomly assigned for each iteration of temperature. When there is no clear majority of a specific spin, the lattice will continually flip random spins to align with their neighbours. This results in areas of magnetic domains where specific areas of the lattice have aligned spins. This is why the first few points of the plot are chaotic - each temperature increment represents a random iteration of the lattice spins that has been given a finite time to equilibrate. As the temperature passes its critical point, the result is the same as the cold start.

```python
## 30x30 lattice with 'High' start but more steps to reach equilibrium

exp1 = model.plots(N=30, start='High',steps=150000)
exp1.lattice()
exp1.show()
```

We can see that if we increase the number of steps (i.e. allowing more time for the lattice to reach equilibrium), the lattice will eventually have all its spins aligned for temperatures before critical tepmerature.

**Question 3: Describe how the spin distribution maps in experiment 3 change with rising temperature.
How does the behaviour you see in the maps correspond to physical properties like magnetization
and susceptibility?**

The spin distributions of the plots below critical temperature show that the lattice has a fairly ordered structure with majority of its spins being aligned. As the the temperature increases, the spontaneous flipping from this thermal energy shows how more and more of these spins cannot be aligned, eventually reaching its chaotic state at critical point. 

We also note that no matter how many steps we define, the lattice will never reach full equibrium for sufficiently large lattices. This seems to be a limitation of the program, rather than a property of the system itself, that arises from the fact that the program uses the Metropolis algorithm that picks a random electron at each step. In the real world, all these electrons should go through this process simulatenously and continuously.

Rising temperature has a direct correlation between both magnetisation and susceptibility. At low temperatures, as most of the spins are aligned, magnetisation is at its max. However, magnetic susceptibility is is 0, until it reaches critical temperature.
